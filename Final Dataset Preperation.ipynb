```python
months=['January','February','March','April','May','June',
        'July','August','September','October','November','December']
```

```python
def is_leap_year(year):
    return (year%4==0 and year%100!=0) or (year%400==0)
```

## Function to convert excel files of months to csv and append dataframes to list

```python
def process_months(year):
    dfs=[]
    for i in months:
        pd.read_excel(f'Data/{year}/AQI_hourly_{year}_{i}_Ashok_Vihar_Delhi_DPCC_{i}_{year}.xlsx', 'Sheet2', index_col=None).to_csv(f'Data/{year}/{year}_{i}.csv',encoding='utf-8',index=False)
        df=pd.read_csv(f'Data/{year}/{year}_{i}.csv')
        dfs.append(df)
    return dfs
```

## Function to create a list of year wise AQI dataframes

```python
def process_year(start_year, end_year):
    temp_year_dfs=[]
    year_list=np.arange(start_year,end_year+1,1).tolist()
    for year in year_list:
        if is_leap_year(year):
            val_list=value_vars_list_leap
        else:
            val_list=value_vars_list
        list=process_months(year)
        final_year_df=pd.concat([process_dataframes(list[i],val_list[i]) for i in range(len(list))],ignore_index=True)
        temp_year_dfs.append(final_year_df)
    return temp_year_dfs
    
```

## Creating AQI dataset from 2019-2023

```python
aqi_list=process_year(2019,2023)
len(aqi_list)
```

```python
aqi_dataset=pd.concat(aqi_list,ignore_index=True)
```
